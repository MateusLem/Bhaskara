#Projeto Bhaskara

##Contexto

Esse projeto surgiu com o objetivo de auxiliar alunos em interpretar equações de 2° grau utilizando a lógica da fórmula de Bhaskara, permitindo que eles vejam os resultados de Δ (delta), de X e Y  do vértice, os pontos em que X corta o eixo Y = 0 e gerem um gráfico que exiba essa função.

##Objetivo
Permitir não apenas calcular os valores de X e Y em uma função de 2° grau, como também saber seu delta e gerar uma imagem que representa a função informada pelo usuário.

##Créditos e informações

* Foram utilizadas as bibliotecas [Numpy](https://numpy.org/) e [Matplotlib](https://matplotlib.org/) para auxiliar nos cálculos e na geração do gráfico, respectivamente.

##Código


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def num():
  """
  Esta função recebe os valores de "a", "b" e "c"

  :param: none

  :return a: float
  :return b: float
  :return c: float
  """

  a = float(input("Informe o valor de 'a': "))
  b = float(input("Informe o valor de 'b': "))
  c = float(input("Informe o valor de 'c': "))
  
  return a,b,c
```

```python
def calcula(a,b,c):
  """
  Esta função calcula os valores x e y do vértice. 
  
  Ela primeiramente calcula o valor de Δ (d)

  Caso esse valor seja menor que 0:
  a função informa que não há raízes reais e encerra o programa

  Caso contrário, ela calcula os dois pontos em que x corta o eixo em que Y = 0 (X1 e X2),
  o extremo de X no vértice (XV) e o ponto de Y que encontra este extremo (YV) 

  :param a: float
  :param b: float
  :param c: float

  :return X1: float
  :return X2: float
  :return XV: float
  :return YV: float
  """

  d = b ** 2 - 4 * a * c

  if d < 0:
    print("Não há raízes reais.")
    return

  else:
    X1 = (-(b)+((d)**(1/2)))/(2*a)
    X2 = (-(b)-((d)**(1/2)))/(2*a)
    XV = -(b)/(2*a)
    YV = -(d)/(4*a)
    print("\n")
    print("="*20)
    print(f"Delta = {d}")
    print("="*20)
    print(f"X1 = {X1:.3f}")
    print("="*20)
    print(f"X2 = {X2:.3f}")
    print("="*20)
    print(f"Xv = {XV:.3f}")
    print("="*20)
    print(f"Yv = {YV:.3f}")
    print("="*20)

  return X1,X2,XV,YV
```

```python
def variacao(x1,x2):
  """
  Esta função calcula a variação entre X1 e X2 
  
  Utilizando a função abs, ela gera um valor int que é a diferença entre ambos 

  Caso esse valor seja menor que 3:
  a função torna a varição = 3

  Caso contrário, exibe o valor dessa variação 

  Por fim, ela retorna o valor da variação

  :param x1: float
  :param x2: float

  :return variacao: int
  """

  variacao = abs(x1 - x2)

  if variacao < 3:
      variacao = 3

  print(f"Variação = {variacao:.3f}")
  print("="*20)
  print("\n")

  return variacao
```

```python
def eixos(x1,x2,a,b,c,eixo_x,eixo_y,zero,var):
  """
  Esta função é responsável por popular as listas que serão usadas para criar o gráfico da variação

  :param x1: float
  :param x2: float
  :param a: float
  :param b: float
  :param c: float
  :param eixo_x: list
  :param eixo_y: list
  :param zero: list
  :param var: int
    
  :return: none
  """
  
  for x in np.arange(x1 - var, x2 + var, var / 100):
    y= a * (x ** 2 ) + b * (x) + c
    eixo_x.append(x)
    eixo_y.append(y)
    zero.append(0.0)
```

```python
def main():
  """
  Esta é a função Main do programa

  :param: none

  :return: none
  """  

  a,b,c=num()

  try:
    x1,x2,xV,yV=calcula(a,b,c)
  except:
    return

  eixo_x = []
  eixo_y = []
  zero = []

  var=variacao(x1,x2)

  eixos(x1,x2,a,b,c,eixo_x,eixo_y,zero,var)

  fig = plt.figure() 
  ax = fig.add_subplot(111)

  plt.plot(eixo_x,eixo_y,color="red") 
  plt.plot(eixo_x,zero,color="black") 

  ax.annotate(str(x1),xy=(x1,0.0)) 
  ax.annotate(str(x2),xy=(x2,0.0)) 

  plt.show()
```

```python
main()
```
